```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\ghoas\OneDrive\Desktop\excel sheets\decisiontree\decision1.csv"
data = pd.read_csv(file_path)
data.head()
```

```python
#convert the string to numpy varaibles using labelEncoder
from sklearn.preprocessing import LabelEncoder

le_encoder = LabelEncoder()
```

```python

data['size_n'] = le_encoder.fit_transform(data['size'])
data['color_n'] = le_encoder.fit_transform(data['color'])
data['target_n'] = le_encoder.fit_transform(data['target'])
```

```python
#drop all string columns
new_data=data.drop(['color','size','target'],axis= 'columns')
new_data
```

```python
data.head()
```

```python
X = new_data.drop(['target_n'],axis='columns')
Y = new_data[["target_n"]]
```

```python
#decision tree split
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X,Y)
```

```python
model.predict([[0,0]])
```

```python
model.score(X,Y)
```

```python
#visualizzze in decision tree
from sklearn.tree import plot_tree

plt.figure(figsize=(15, 10))
plot_tree(model, feature_names=['size_n', 'color_n'],class_names=data['target'].unique(), filled=True, rounded=True, fontsize=8)
plt.title('Decision Tree Visualization', fontsize=16)
plt.show()
```
