```python
# import necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris= load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
# cretae a dataFrame of the iris dataset

df= pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
# add target column 0,1,2 are setosa', 'versicolor', 'virginica
df['target']= iris.target
df.head()
```

```python
iris.target_names
```

```python
#means this lies in the versicolor category
df[df.target == 1].head()
```

```python
#add new column flower name from target column that are 0 1 2

#if any number have to be changed to respective number use pandas apply function
#using lmbda to get the value of each ctarget in x the futher check the number in iris.target_names , asign respective flower name and add in its row 
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
#visualize
#lets spread 3 species into 3 dataframe

df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]

#
```

```python
#scatter plot these now for sepal

plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='green',marker="+")
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker="o")
plt.scatter(df2['sepal length (cm)'],
            df2['sepal width (cm)'], color='red', marker="s")
plt.xlabel("Length of sepal")
plt.ylabel("Width of sepal")
```

```python
# scatter plot these now for petal

plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],
            color='green', marker="+")
plt.scatter(df1['petal length (cm)'],
            df1['petal width (cm)'], color='blue', marker="o")
plt.scatter(df2['petal length (cm)'],
            df2['petal width (cm)'], color='red', marker="s")
plt.xlabel("Length of petal")
plt.ylabel("Width of petal")
```

```python
#lets split and train

X=df.drop(['target','flower_name'],axis='columns')
y= df.target
```

```python
X_train,X_test,y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
# lets train
from sklearn.svm import SVC

# model = SVC(C=100) #best gives 1 score 
# model=SVC(gamma=100)

#use kernal 
model = SVC()

#can change the scale to gamma, c
```
