```python
X=data
y= data.target
```

```python
#lets convert to scaler

from sklearn.preprocessing import StandardScaler
st= StandardScaler()
```

```python
X_scaled = st.fit_transform(X)
X_scaled.shape
```

```python
X_scaled
```

```python
new_data = pd.DataFrame(X_scaled)
```

```python
new_data.head()
```

```python
#split 
y= data.target

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=10)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression

lg= LogisticRegression()
lg.fit(X_train,y_train)
```

```python
lg.score(X_test,y_test)
```

```python
X_scaled.shape
```

```python
#PCA

from sklearn.decomposition import PCA

pca = PCA(0.95) #whta it means #retain 95% of useful features
X_pca = pca.fit_transform(X)
X_pca.shape

#got only 29 on 65 redit opf unnecessary
```

```python
X_pca
```

```python
pca.explained_variance_ratio_

# saying my first column is capturing 14% of variation, 14% of useful variations
```

```python
#basically columns number
pca.n_components_
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_pca,y , test_size = 0.2 , random_state=30)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
#no of component is 2 menas only 2 main components
pca= PCA(n_components = 2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
X_train, X_test,y_train,y_test = train_test_split(X_pca,y , test_size = 0.2 , random_state=30)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
