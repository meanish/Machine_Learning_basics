```python
import numpy as np
import pandas as pd

path = r'C:\Users\ghoas\OneDrive\Desktop\excel sheets\heart.csv'
data = pd.read_csv(path)
data.head()
```

```python
data['ChestPainType'].unique
```

```python
sex_dummy=pd.get_dummies(data.Sex,drop_first=True)
chest_paindummy = pd.get_dummies(data.ChestPainType,drop_first=True) 
ecg_dummmy = pd.get_dummies(data.RestingECG, drop_first=True) 
exe_dummmy = pd.get_dummies(data.ExerciseAngina, drop_first=True)
st_dummmy = pd.get_dummies(data.ST_Slope, drop_first=True)
```

```python
data = pd.concat([data,sex_dummy,chest_paindummy,ecg_dummmy,exe_dummmy,st_dummmy],axis='columns')
```

```python
data.head()
```

```python
data = data.drop(['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina','ST_Slope'],axis='columns')
```

```python
data.describe()
```

```python

# Treat Outliers is the outliers 
data[data.Cholesterol > (data.Cholesterol.mean()+3*data.Cholesterol.std())]
```

```python
data[data.Oldpeak > (data.Oldpeak.mean()+3*data.Oldpeak.std())]
```

```python
data2 = data1[data1.Oldpeak <= (data1.Oldpeak.mean()+3*data1.Oldpeak.std())]
data2.shape
```

```python
data.head()
```

```python
data3 = data2[data2.RestingBP <= (data2.RestingBP.mean()+3*data2.RestingBP.std())]
data3.shape
```

```python
X = data3.drop(['HeartDisease'], axis='columns')
y=data3.HeartDisease
```
