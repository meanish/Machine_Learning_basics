```python
air_quality_long['date_time'] = pd.to_datetime(air_quality_long['date.utc'])
```

```python

air_quality_long
```

```python
air_quality_long = air_quality_long.rename(columns={"date.utc": "datetime"})
```

```python
#convert to datetime beacause intial is a string value cant do date time features as find month yera etc
air_quality_long['datetime'] = pd.to_datetime(air_quality_long['datetime'])
```

```python
air_quality_long["datetime"]
```

```python
#lets a a new column where only months fromthe timestamp

air_quality_long['month']= air_quality_long['datetime'].dt.month
```

```python
air_quality_long
```

```python
onlyno = air_quality_long[air_quality_long['parameter'] == "no2"]
onlyno
```

```python
#only find the no2 values with the weeks number

weekValue = onlyno.groupby([onlyno['datetime'].dt.weekday, 'location'])[
    'value'].mean()
weekValue
```

```python
# representation with hours value
fig, axs = plt.subplots(figsize=(12, 4))

onlyno.groupby([onlyno['datetime'].dt.hour])[
    'value'].mean().plot(kind='bar', rot=0, ax=axs)

axs.set_xlabel("Hour of the day")
axs.set_ylabel ("Avg. value")
plt.show()
```
