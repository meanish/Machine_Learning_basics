```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris= load_iris()
dir(iris)
```

```python
data= pd.DataFrame(iris.data, columns=iris.feature_names)
data.head()
```

```python
#addd atget 
data['target']= iris.target
```

```python
data.head()
```

```python
#conmnvert target to respective names

data['target'] = data.target.apply(lambda x: iris.target_names[x])
data.head()
```

```python
data.head
```

```python
# we ahve cross val to check whch is the bes 

from sklearn.model_selection import  cross_val_score

from sklearn.svm import SVC

#do sv,m method for few parameters lambda, c etc


```

```python
val=cross_val_score(SVC(kernel='linear', C=10, gamma='auto'),iris.data,data.target,cv=5)
avg= val.mean()
avg
```

```python
val = cross_val_score(SVC(kernel='sigmoid', C=10, gamma='auto'),
                iris.data, data.target, cv=5)
avg = val.mean()
avg
```

```python
val = cross_val_score(SVC(kernel='rbf', C=10, gamma='auto'),
                iris.data, data.target, cv=5)
avg = val.mean()
avg
```

```python
# To know tjhe optimal hypermeter need to do manually

#or use for loop

kernels = [ 'linear','sigmoid','rbf']
C=[1,10,20]
avg_scores= { }

for kval in kernels:
    for c in C:
        val = cross_val_score(SVC(kernel=kval, C=c, gamma='auto'),
                          iris.data, data.target, cv=5)
        avg = val.mean()
        print("The avg for {} is {}".format(kval,avg))
   
```

```python
# thsi is also lengthy need to manaully give C 

#here comes gridSearchCV

from sklearn.model_selection import GridSearchCV

Gsv= GridSearchCV(SVC(gamma='auto'),{
    #parameter grid
    'C':[1,10,20],
    'kernel':['linear','rbf','sigmoid'],
},cv=5, return_train_score=False)  #False because we dont neeed parameter that it retrurn


Gsv.fit(iris.data,data.target)
Gsv.cv_results_

```

```python
# All with the answer and mewan, 
#to study the result pandas provided a library

df = pd.DataFrame(Gsv.cv_results_)
df
```

```python
# but theres is an issue here what is c value is in millions
# millions of permutaion and computation has to be run
# increase in omputation cost
# nCm is te computation where ab=ba
# cpm is permutation where ab != ba


# to tackle the issue introduced random search cv
from sklearn.model_selection import RandomizedSearchCV

rs = RandomizedSearchCV(SVC(gamma='auto'), {
    # parameter grid
    'C': [1, 10, 20],
    'kernel': ['linear', 'rbf', 'sigmoid'],
}, cv=5, return_train_score=False,
    n_iter=2)

#n_iter is the number of ietration before if there is a rbf and C=[10,0,20] the iteration runs for all 3 C'set
#but now just 2 random
```

```python
rs.fit(iris.data,data.target)
```

```python
val= pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
val
```

```python
# how to choose the best model for optimal performance

# what I want to test RF, LogisticRegression, svc etc etc
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier




model_params = {
    'svm': {
        'model': SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf','linear', 'sigmoid']
        }
    },
    'Logistic_regression': {
        'model':LogisticRegression(),
        'params':{
            'C': [1,5,10]
        }

    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    }

}
```

```python
# use RandomSeacrch hypermeter Cv

scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'],
                       cv=5, return_train_score=False)
    clf.fit(iris.data, iris.target)
    scores.append(
        {
            'model': model_name,
            'best_score': clf.best_score_,
            'best_params': clf.best_params_
        }
    )

print('The score of all is {}'.format(scores))    
```
