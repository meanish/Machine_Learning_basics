```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = r"C:\Users\ghoas\OneDrive\Desktop\excel sheets\Linear1.csv"
```

```python
home = pd.read_csv(data)

dir(linear_model)
```

```python
home
```

```python
fig, axs = plt.subplots(figsize=(12, 4))
axs.set_xlabel("Area of homes")
axs.set_ylabel("Price of homes")
plt.scatter(home.area, home.price, color="red", marker="+")
# plt.plot(home.area, home.price, color='blue', linewidth=3) for plotting each points
```

```python
# Declation X and Y

X = home['area'].values.reshape(-1,1)    # independent varaible
y = home['price'].values.reshape(-1,1)  # dependent varaibles
#the main reasion for using reshape is converting 1D to @d say thee are 5 elements -1 makes it 5 rows, 1 makes it sya atleast 1 col.
#  Converted to 2D beacuse they are important for all macjhine leari=ning libaruies
X 
```

```python
# linear regression euation create
reg = linear_model.LinearRegression()
reg.fit(X, y)
```

```python
reg.predict([[3300]])
```
