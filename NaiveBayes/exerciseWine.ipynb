```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_wine

wine = load_wine()
```

```python
(dir(wine))
```

```python
data = pd.DataFrame(wine.data,columns = wine.feature_names)
data.head()
```

```python
data['target'] = wine.target
```

```python
data.head()
```

```python
#split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train, y_test  = train_test_split(wine.data,data.target,test_size=0.2,random_state=42)
```

```python
#train with GaussianNb
from sklearn.naive_bayes import GaussianNB
gB = GaussianNB()
gB.fit(X_train,y_train)
```

```python
gB.score(X_test,y_test)
```

```python
# train with GaussianNb
from sklearn.naive_bayes import MultinomialNB
Mn = MultinomialNB()
Mn.fit(X_train, y_train)
```

```python
Mn.score(X_test, y_test)
```

```python
# lets see frokm cross validation

from sklearn.model_selection import cross_val_score

score_gB = cross_val_score(GaussianNB(), wine.data,data.target)
score_M = cross_val_score(MultinomialNB(), wine.data,data.target)

print("The avg of both the test Gb and Mn are {:2f} and {:2f} respectvely".format(score_gB.mean(),score_M.mean()))
```
